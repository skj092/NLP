**Compare the lexical diversity scores for humor and romance fiction in Ta-ble 1-1. Which genre is more lexically diverse?**

```python
from nltk.corpus import brown

genre = brown.categories()

tokens = [len(brown.words(categories=i)) for i in brown.categories()]

types = [len(set(brown.words(categories=i))) for i in brown.categories()]

data = {'Genre': genre, "Tokens": tokens, 'Types': types}
```

```python
import pandas as pd 
df = pd.DataFrame(data)

df['Lexical diversity'] = df['Tokens']/df['Types']

df
```
